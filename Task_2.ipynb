```python

import random
import time

def bfs(lis, goal):
    if not lis:
        return None

    queue = [(lis, 0)]
    visited = set()

    while queue:
        current, depth = queue.pop(0)
        visited.add(tuple(current))

        if current[-220:] == goal:
            return depth

        for i in range(len(current)):
            new_list = current[:i] + current[i+1:]
            if tuple(new_list) not in visited:
                queue.append((new_list, depth + 1))

    return -1

def dfs(lis, goal):
    if not lis:
        return None

    stack = [(lis, 0)]
    visited = set()

    while stack:
        current, depth = stack.pop()
        visited.add(tuple(current))

        if current[-220:] == goal:
            return depth

        for i in range(len(current)):
            new_list = current[:i] + current[i+1:]
            if tuple(new_list) not in visited:
                stack.append((new_list, depth + 1))

    return -1

# Example usage
random.seed(42)  # Set random seed for reproducibility

# Generate a random set of numbers
total_len = 1000
lis = [random.randint(1, 1000) for _ in range(total_len)]

goal = lis[total_len - 220:]

# Measure BFS execution time
start_time = time.time()
bfs_depth = bfs(lis, goal)
end_time = time.time()
execution_time = end_time - start_time
print("BFS Execution Time:", execution_time)
print("BFS Depth:", bfs_depth)

# Measure DFS execution time
start_time = time.time()
dfs_depth = dfs(lis, goal)
end_time = time.time()
execution_time = end_time - start_time
print("DFS Execution Time:", execution_time)
print("DFS Depth:", dfs_depth)
```
